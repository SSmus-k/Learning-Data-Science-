```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

#  given data
sample_mean = 3000
population_mean = 2850
sigma = 25
n = 100
alpha = 0.05 # significance level 95%

SE = sigma / np.sqrt(n)  # standard error
print (SE)

z = (sample_mean - population_mean) / SE  # z-score
print("Z-score:", z)

z_critical = stats.norm.ppf(1 - alpha/2)  # critical z-value for one-tailed test
print("Critical Z-value:", z_critical)

if abs(z) > z_critical:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
sample_mean = 27436
population_mean = 28000
n = 40

sigma = 1348  
alpha = 0.01 # significance level 99%


SE = sigma / np.sqrt(n)  # standard error
print ("Standard Error:",SE)

z = (sample_mean - population_mean) / SE  # z-score
print("Z-score:", z)

z_critical = stats.norm.ppf(alpha)  # critical z-value for one-tailed test
print("Critical Z-value:", z_critical)

# p value
p_value = stats.norm.cdf(z)
print("P-value:", p_value)

rejecth0 = p_value < alpha
print("Reject null hypothesis:", rejecth0)

if abs(z) > z_critical:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

```

```python
import numpy as np
from scipy import stats

data = np.array([3005, 2935, 2995, 2965, 3905, 2395, 2905])
mu_0 = 3000
alpha = 0.05

n = len(data)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)

t_statistic = (sample_mean - mu_0) / (sample_std / np.sqrt(n))
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
print("Critical t-value:", t_critical)
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=n-1))

print("degrees of freedom:", n-1)
print("t Statistic:", t_statistic)
print("p-value:", p_value)

if t_statistic > t_critical:
    print("Decision: Reject the null hypothesis")
    print("Conclusion: The mean is significantly different from 3000")
else:
    print("Decision: Fail to reject the null hypothesis")
    print("Conclusion: No significant difference from mean = 3000")


```
